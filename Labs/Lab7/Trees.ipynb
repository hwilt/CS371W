```python
import numpy as np
import matplotlib.pyplot as plt

class TreeNode(object):
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None
        self.inorder_pos = 0
        self.height = 0
    
    def add(self, key):
        ret = self
        if key < self.key:
            if self.left:
                self.left = self.left.add(key)
            else:
                self.left = TreeNode(key)
        elif key > self.key:
            if self.right:
                self.right = self.right.add(key)
            else:
                self.right = TreeNode(key)
        return ret
    
    def inorder(self, num, key_list):
        """
        Parameters
        ----------
        num: list
            List of a single element which keeps 
            track of the number I'm at
        """
        if self.left:
            self.left.inorder(num, key_list)
        self.inorder_pos = num[0]
        key_list.append(self.key)
        num[0] += 1
        if self.right:
            self.right.inorder(num, key_list)
    
    def draw(self, y):
        x = self.inorder_pos
        plt.scatter([x], [y], 50, 'k')
        plt.text(x+0.2, y, "{}".format(self.key))
        plt.text(x+0.7, y, "({})".format(self.height), color='r')
        y_next = y-1
        if self.left:
            x_next = self.left.inorder_pos
            plt.plot([x, x_next], [y, y_next])
            self.left.draw(y_next)
        if self.right:
            x_next = self.right.inorder_pos
            plt.plot([x, x_next], [y, y_next])
            self.right.draw(y_next)
        
        
class BinaryTree(object):
    def __init__(self):
        self.root = None
    
    def inorder(self):
        key_list = []
        if self.root:
            self.root.inorder([0], key_list)
        return key_list
    
    def draw(self):
        self.inorder()
        if self.root:
            self.root.draw(0)
    
    def add(self, key):
        if self.root:
            self.root = self.root.add(key)
        else:
            self.root = TreeNode(key)

        self.updateheight(self.root)

        if self.getbalance(self.root) > 1:
            if self.getbalance(self.root.left) > 0:
                self.root = self.rotate_left(self.root)
            else:
                self.root = self.rotate_right(self.root)
        elif self.getbalance(self.root) < -1:
            if self.getbalance(self.root.right) < 0:
                self.root = self.rotate_right(self.root)
            else:
                self.root = self.rotate_left(self.root)

    def rotate_left(self, root):
        new_root = root.right
        root.right = new_root.left
        new_root.left = root
        self.updateheight(root)
        self.updateheight(new_root)
        return new_root

    def rotate_right(self, root):
        new_root = root.left
        root.left = new_root.right
        new_root.right = root
        self.updateheight(root)
        self.updateheight(new_root)
        return new_root


    def updateheight(self, root):
        if root:
            root.height = 1 + max(self.getheight(root.left), self.getheight(root.right))


    def getheight(self, node):
        if not node:
            return 0
        else:
            return node.height
    
    def getbalance(self, root):
        if self.root:
            return 0
        else:
            return self.getheight(root.left) - self.getheight(root.right)

def make_tree():
    T = BinaryTree()
    for val in [10, 7, 16, 3, 9, 11, 20, 14, 17, 13, 12, 6, 5, 17, 15, 8, 4]:
        T.add(val)
    return T

T = make_tree()
plt.figure()
T.draw()
```
